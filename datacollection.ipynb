```python
import requests

url3 = 'https://en.wikipedia.org/wiki/Health_care_systems_by_country'
response = requests.get(url3)
content = response.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')
heading = soup.find('span', id='Countries_with_universal_government-funded_health_system')
print(heading)
```

```python
countries = [] #empty list
#list with only the countries and no superscript numbers
countries = [a.text for a in heading.find_next('ul').find_all('a') if not any(c.isdigit() for c in a.text)]
print(countries)
```

```python
countries_list = le['Countries'].tolist()
mc = set(countries) - set(countries_list)
print(f"Countries in un but not in le: {mc}")
len(mc)
```

```python
values_to_remove = ['Taiwan', 'Faroe Islands', 'Åland Islands', 'North Korea', 'San Marino', 'Greenland']
for value in values_to_remove:
    while value in countries:
        countries.remove(value)
```

```python
countries_list = le['Countries'].tolist()
mc = set(countries) - set(countries_list)
print(f"Countries in un but not in le: {mc}")
len(mc)
```

```python
le['Universal Health Care'] = 0
le['Universal Health Care'] = le['Countries'].apply(lambda x: 1 if x in countries else 0)
```

```python
le = le.loc[:, ['Countries','All', 'Universal Health Care']]
le=le.iloc[:, [0, 1, -1]]
le = le.drop(le[le['Countries'] == 'Albania'].index) #no health expenditure data for 2019
exp = exp.drop(exp[exp['Location'] == 'Albania'].index)
le = le.drop(le[le['Countries'] == 'Saudi Arabia'].index) #no health expenditure data for 2019
exp = exp.drop(exp[exp['Location'] == 'Saudi Arabia'].index)
exp=exp.iloc[:,[0,2]]
```

```python
df = pd.merge(exp, le, left_on='Location', right_on='Countries')
df.rename(columns={'2019': '2019 Health Expenditure Per Capita', 'All': '2019 Life Expectancy at Birth', 'Universal Health Care': 'UHC'}, inplace=True)
# drop the "Countries" column
df = df.drop(labels='Countries', axis=1)
df.head(60)
```
